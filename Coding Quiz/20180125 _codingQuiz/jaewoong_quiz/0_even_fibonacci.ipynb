```python
% % timeit
fib = [1, 1]
limit = 4000000
i = 2
sum = 0

while fib[i - 2] + fib[i - 1] < limit:
    fib.append(fib[i - 2] + fib[i - 1])

    if fib[-1] > limit:
        break

    elif fib[-1] % 2 == 0:
        sum += fib[-1]

    i += 1

# print(sum)
```
